## Linear Regression

```python
# house Pricing Dataset
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import fetch_california_housing
```

```python
df = fetch_california_housing()
```

```python
df
```

```python
dataset = pd.DataFrame(df.data)
dataset.columns = df.feature_names
dataset.head()
```

```python
dataset["Price"] = df.target
dataset.head()
```

```python
## Dividing the dataset into independent and dependent features

X = dataset.iloc[:,:-1]  #independent feature
y = dataset.iloc[:,-1]  #dependent feature
```

```python
X.head()
```

```python
y.head()
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

lin_reg = LinearRegression()
lin_reg.fit(X,y)

# Cross Validation
mse = cross_val_score(lin_reg,X,y,scoring='neg_mean_squared_error', cv=5)
mean_mse = np.mean(mse)

print(mean_mse)
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV #this will help with hyper parameter tuning

ridge = Ridge()

params = {'alpha': [1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,50,55,60,100]}

ridge_regressor = GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X,y)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV #this will help with hyper parameter tuning

lasso = Lasso()

params = {'alpha': [1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,50,55,60,100]}

lasso_regressor = GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(X,y)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
## Dataset

from sklearn.datasets import load_breast_cancer
df = load_breast_cancer()

# Independent features

X = pd.DataFrame(df['data'],columns=df['feature_names'])
```

```python
X.head()
```

```python
## Dependent Feature

y = pd.DataFrame(df['target'],columns =["Target"] )
```

```python
y.head()
```

```python
# To check if the data is balanced or imbalanced

y["Target"].value_counts()
```

```python
## Train Test split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42) # this means test data will be 33% & train data will be 77%
```

```python
params = [{'C':[1,5,10]}, {'max_iter': [100,150]}]
```

```python
model1 = LogisticRegression(C=100, max_iter=100)
```

```python
model = GridSearchCV(model1, param_grid=params, scoring='f1',cv=5)
```

```python
model.fit(X_train,y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
y_pred = model.predict(X_test)
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
